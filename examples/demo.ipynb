```python
%load_ext autoreload
%autoreload 2

import sys
sys.path.append('../')

import matplotlib.pyplot as plt

from time import time

from sodgen.config import Config
from sodgen.generate import generator
```

# Create config with desired results

```python
cfg = Config()
cfg.lang = 'zh-CN'
cfg.fonts_dir = './fonts_dir'
cfg.corpus_path = './test_english_corpus.csv'
cfg.size = (512, 512)
cfg.stroke_width = (1, 2)
cfg.font_size = (14, 22)
cfg.target_text_number = (4, 6)

cfg.max_line_length = 20
```

# Generate images

```python
num = 16

generator = generator(config=cfg)

start = time()
ims = generator.generate_images(num)
print(f'{time() - start} seconds to generate {num} images')
```

# Show results

```python
plt.figure(figsize=(20, 20))

for i in range(1, num + 1):
    plt.subplot(4, 4, i)
    plt.imshow(ims[i - 1].image)
```
